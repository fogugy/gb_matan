```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sin, cos
%matplotlib inline
```

```python
def four(x, k=6):
    res = (pi**2)/3
    
    for n in range(1, k+1):
        res += 4*((-1)**n)*cos(n*x)/n**2
        
    return res
```

```python
x_ = np.arange(-pi, pi, pi/100)
y_f = list(map(lambda x: x**2, x_))
y_fourier = list(map(lambda x: four(x), x_))
```

```python
plt.plot(x_, y_f, color='red');
plt.plot(x_, y_fourier, color='green');
```
