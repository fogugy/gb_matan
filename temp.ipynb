```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, pi, factorial, sinh, sin, cos

%matplotlib inline
```

```python
x_ = np.arange(-pi, pi, 0.01)
y_ = list(map(lambda x: exp(x), x_))
```

```python
def mcloren(x, k=10):
    res = 1
    for n in range(1, k+1):
        res += (x**n)/factorial(n)
    return res

def four(x, k=120):
    res = sinh(pi)/pi
    for n in range(1, k+1):
        res += (2*sinh(pi)*cos(pi*n))*(cos(n*x)-n*sin(n*x))/(pi*(n**2 + 1))
    return res
```

```python
y_mc = list(map(mcloren, x_))
y_f = list(map(four, x_))
```

```python
plt.plot(x_, y_, color='red', linewidth=6);
plt.plot(x_, y_mc, color='yellow', linewidth=4);
plt.plot(x_, y_f, color='black');
```
